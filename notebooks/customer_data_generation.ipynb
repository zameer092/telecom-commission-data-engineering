```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
def generate_customer(n=1000):
    data = []
    
    for i in range(1, n+1):
        data.append({
            "customer_id": f"C{i}",
            "customer_name": f"Customer_{i}",
            "city": random.choice(["Mumbai", "Delhi", "Pune", "Bangalore"]),
            "state": random.choice(["MH", "DL", "KA", "TN"])
        })
    
    return pd.DataFrame(data)
 
df = generate_customer(1000)
 
df.head()
```

```python
file_name = "customer_20260429.csv"
df.to_csv(file_name, index=False)
 
print("Customer file created")
```

```python
from google.cloud import storage
 
client = storage.Client()
bucket = client.bucket("telecom-commission-data")
 
blob = bucket.blob(f"landing/customer/{file_name}")
blob.upload_from_filename(file_name)
 
print("Uploaded to GCS")
```
