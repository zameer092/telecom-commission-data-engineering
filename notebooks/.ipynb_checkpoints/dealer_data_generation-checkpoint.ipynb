```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
def generate_dealer(n=100):
    data = []
    
    for i in range(1, n+1):
        dealer = {
            "dealer_id": f"D{i}",
            "dealer_name": f"Dealer_{i}",
            "region": random.choice(["North", "South", "East", "West"]),
            "join_date": datetime.now() - timedelta(days=random.randint(100, 1000))
        }
        data.append(dealer)
    
    df = pd.DataFrame(data)
    return df
 
df_dealer = generate_dealer(100)
 
df_dealer.head(20)
```

```python
file_name = "dealer_20260428.csv"
df_dealer.to_csv(file_name, index=False)
```

```python
from google.cloud import storage
 
client = storage.Client()
bucket = client.bucket("telecom-commission-data")
 
blob = bucket.blob(f"landing/dealer/{file_name}")
blob.upload_from_filename(file_name)
 
print("File uploaded successfully")
```
