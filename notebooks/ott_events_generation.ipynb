```python
import pandas as pd
import random
from datetime import datetime, timedelta
 
def generate_ott(n=1500):
    data = []
    
    ott_types = ["Netflix", "Prime", "Hotstar"]
    
    for i in range(1, n+1):
        data.append({
            "subscription_id": f"S{i}",
            "customer_id": f"C{random.randint(1,1000)}",
            "dealer_id": f"D{random.randint(1,100)}",
            "agent_id": f"AG{random.randint(1,50)}",
            "pos_id": f"P{random.randint(1,200)}",
            "ott_type": random.choice(ott_types),
            "subscription_type": random.choice(["new","renewal"]),
            "subscription_date": datetime.now() - timedelta(days=random.randint(1,30))
        })
    
    return pd.DataFrame(data)
 
df = generate_ott(1500)
 
df.head()
```

```python
file_name = "ott_20260429.csv"
df.to_csv(file_name, index=False)
```

```python
from google.cloud import storage
 
client = storage.Client()
bucket = client.bucket("telecom-commission-data")
 
blob = bucket.blob(f"landing/ott/{file_name}")
blob.upload_from_filename(file_name)
 
print("OTT uploaded")
```
