```python
import pandas as pd
import numpy
import random
from datetime import datetime, timedelta
```

```python
def generate_sim_events(n=2000):
    data = []
    
    for i in range(1, n+1):
        data.append({
            "activation_id": f"A{i}",
            "customer_id": f"C{random.randint(1,1000)}",
            "dealer_id": f"D{random.randint(1,100)}",
            "agent_id": f"AG{random.randint(1,50)}",
            "pos_id": f"P{random.randint(1,200)}",
            "activation_date": datetime.now() - timedelta(days=random.randint(1,30)),
            "plan_type": random.choice(["prepaid","postpaid"]),
            "event_type": random.choice(["activation","deactivation","upgrade"])
        })
    
    return pd.DataFrame(data)
 
df = generate_sim_events(2000)
 
df.head()
```

```python
file_name = "sim_20260429.csv"
df.to_csv(file_name, index=False)
 
print("SIM file created")
```

```python
from google.cloud import storage
 
client = storage.Client()
bucket = client.bucket("telecom-commission-data")
 
blob = bucket.blob(f"landing/sim/{file_name}")
blob.upload_from_filename(file_name)
 
print("SIM uploaded")
```
