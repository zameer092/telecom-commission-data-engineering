```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
def generate_recharge(n=3000):
    data = []
    
    for i in range(1, n+1):
        data.append({
            "recharge_id": f"R{i}",
            "customer_id": f"C{random.randint(1,1000)}",
            "dealer_id": f"D{random.randint(1,100)}",
            "agent_id": f"AG{random.randint(1,50)}",
            "pos_id": f"P{random.randint(1,200)}",
            "amount": random.choice([50,100,200,300,500]),
            "recharge_type": random.choice(["normal","addon"]),
            "recharge_date": datetime.now() - timedelta(days=random.randint(1,30))
        })
    
    return pd.DataFrame(data)
 
df = generate_recharge(3000)
 
df.head()
```

```python
file_name = "recharge_20260429.csv"
df.to_csv(file_name, index=False)
 
print("Recharge file created")
```

```python
from google.cloud import storage
 
client = storage.Client()
bucket = client.bucket("telecom-commission-data")
 
blob = bucket.blob(f"landing/recharge/{file_name}")
blob.upload_from_filename(file_name)
 
print("Recharge uploaded")
```
