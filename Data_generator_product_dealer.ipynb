```python
import pandas as pd
from google.cloud import storage
 
BUCKET = "telecom-commission-data"
PATH = "landing/dealer_events.csv"
 
data = [
    ["E001","D001","C001","SIM_ACTIVATION","P001",299,"2026-05-01"],
    ["E002","D002","C002","RECHARGE","P002",399,"2026-05-02"],
    ["E003","D003","C003","TOPUP","P004",100,"2026-05-03"],
    ["E004","D001","C004","ADDON","P007",50,"2026-05-04"],
    ["E005","D002","C005","RECHARGE","P003",499,"2026-05-05"],
    ["E006","D003","C001","TOPUP","P005",200,"2026-05-06"],
    ["E007","D001","C002","SIM_ACTIVATION","P002",399,"2026-05-07"],
    ["E008","D002","C003","ADDON","P007",50,"2026-05-08"],
    ["E009","D003","C004","RECHARGE","P001",299,"2026-05-09"],
    ["E010","D001","C005","TOPUP","P006",500,"2026-05-10"]
]
 
df = pd.DataFrame(data, columns=[
    "event_id","dealer_id","customer_id","event_type",
    "product_id","amount","event_date"
])
 
# Save clean CSV
df.to_csv("dealer_events.csv", index=False)
 
# Upload to GCS
client = storage.Client()
bucket = client.bucket(BUCKET)
blob = bucket.blob(PATH)
blob.upload_from_filename("dealer_events.csv")
 
print("10 clean dealer records uploaded")
 
```
